```python
def convert_k_to_int(val):
    if isinstance(val, str) and 'k' in val.lower():
        return int(float(val.lower().replace('k', '')) * 1000)
    try:
        return int(val)
    except:
        return None


data['num_seller_ratings'] = data['num_seller_ratings'].apply(convert_k_to_int)
```
